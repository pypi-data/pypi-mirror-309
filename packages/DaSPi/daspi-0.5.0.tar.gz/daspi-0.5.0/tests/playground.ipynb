```python
source = pd.DataFrame({
    'A': [1, 0, 1, 1, 0, 1],
    'B': [-1, 1, 0, -1, 1, 0],
    'C': [3.1, 6.1, 9.1, 12.0, 15.0, 18.1],
    'Target': [11, 19, 30, 42, 49, 50]})
target = 'Target'
categorical = ['A', 'B', 'C']
continuous = []
alpha = 0.05
lm = daspi.LinearModel(source, target, categorical, continuous, alpha, order=2)
gofs = list(lm.recursive_feature_elimination())
lm
```

```python
import numpy as np
import pandas as pd

dm = pd.DataFrame({
        'Intercept': [1, 1, 1, 1, 1, 1],
        'A': [-1, 1, 1, -1, -1, 1],
        'B': [0, -1, 0, -1, 1, 1],
        'AB': [0, -1, 0, 1, -1, 1],
        'Center': [1, 0, 1, 0, 0, 0]})

X = dm.values
# P = X(X'X)^(-1)X'
P = np.dot(X, np.dot(np.linalg.inv(np.dot(X.T, X)), X.T))
p_diag = np.diag(P)
p_diag
```
